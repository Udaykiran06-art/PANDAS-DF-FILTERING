```python
#Filtering DataFrame by using relational operator: single condition
import pandas as pd
df1 = pd.read_csv("sales4.csv")
con1 = df1['Product_Cost']>65000
df2 = df1[con1]
print(df2)
```

```python
#Filtering DataFrame by using relational operator : single condition
import pandas as pd
df1 = pd.read_csv("sales4.csv")
con1 = df1['Product_Cost']>70000
df2 = df1[con1]
print(df2)
```

```python
#Filtering DataFrame by using relational operator : Multiple conditions
import pandas as pd
df1 = pd.read_csv("sales4.csv")

con1 = df1['Product_Cost']>50000
con2 = df1['Product_Cost']>60000

df2 = df1[con1&con2]
print(df2)
```

```python
#Filtering DataFrame by using relational operator:Multiple conditions
import pandas as pd
df1 = pd.read_csv("sales4.csv")
con1 = df1.Product_Name == "iPhone 11"
con2 = df1.Customer_Name == "Nireekshan"
df2 = df1[con1&con2]
print(df2)
```

```python
#Filtering DataFrame by using relational operator : Multiple conditions
import pandas as pd
df1 = pd.read_csv("sales4.csv")
con1 = df1.Product_Name == 'iPhone 11'
con2 = df1.Customer_Name == 'Shahid'
df2 = df1[con1&con2]
print(df2)
```

```python
#Filtering DataFrame by using loc indexer
import pandas as pd
df1 = pd.read_csv("sales4.csv")
con1 = df1.Product_Name == 'iPhone 11'
con2 = df1.Customer_Name == 'Shahid'
df2 = df1.loc[con1&con2]
print(df2)
```

```python
#Filtering DataFrame by using iloc indexer
import pandas as pd
df1 = pd.read_csv("sales4.csv")
df2 = df1.loc[:5,]
print(df2)
```

```python
#Filtering DataFrame by using loc indexer
import pandas as pd
df1 = pd.read_csv("sales4.csv")
rows = df1.index[0:]
cols = ["Product_Name","Customer_Id"]
df2 = df1.loc[rows,cols]
print(df2)
```

```python
#Filtering DataFrame by using loc indexer
import pandas as pd
df1 = pd.read_csv("sales4.csv")
a = df1.Product_Name == "LG Washing Machine"
b = df1.Customer_Id == 1
c = a | b
df2 = df1.loc[c]
print(df2)
```

```python
#Filtering DataFrame by using loc isin() method
import pandas as pd
df1 = pd.read_csv("sales4.csv")
a = ["Macbook Pro Laptop"]
b = df1.Product_Name.isin(a)
df2 = df1[b]
print(df2)
```

```python
#Filtering DataFrame by using by using loc isin()method
import pandas as pd
df1 = pd.read_csv("sales4.csv")
a = ["34in Ulterwide Monitor","Macbook Pro Laptop"]
b = df1.Product_Name.isin(a)
df2 = df1[b]
print(df2)
```

```python
#Creating a DataFrame
import pandas as pd
import numpy as np
data = [
    ['Shahid',21,40000],
    ['Nireekshan',22,20000],
    ['Veeru',45,90000],
    ['Sumanth',20,95000],
    [np.nan,2,99000],
    ['Prasad',1,41000]
]
c = ['Name','Age','Salary']
df1 = pd.DataFrame(data,columns=c)
print(df1)
```

```python
#notnull()method
import pandas as pd
import numpy as np
data = [
    ['Shahid',21,40000],
    ['Nireekshan',22,20000],
    ['Veeru',45,90000],
    ['Sumanth',20,95000],
    [np.nan,2,99000],
    ['Prasad',1,41000]
]
c = ['Name','Age','Salary']
df1 = pd.DataFrame(data,columns=c)
d = df1.Name.notnull()
df2 = df1[d]
print(df1)
print()
print(df2)
```
